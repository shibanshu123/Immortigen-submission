```python
import pandas as pd
import numpy as np
```

```python
footfall_df = pd.read_csv("footfall_history.csv")
menu_df = pd.read_csv("menu.csv")
weather_df = pd.read_csv("weather.csv")
calendar_df = pd.read_csv("calendar.csv")
```

```python
footfall_df.describe()
footfall_df.isnull().sum()
footfall_df['timestamp'] = pd.to_datetime(footfall_df['date'] + ' ' + footfall_df['time'])
footfall_df.drop([ 'time'], axis=1, inplace=True)
footfall_df['timestamp']=pd.to_datetime(footfall_df['timestamp'])
footfall_df['date']=pd.to_datetime(footfall_df['date'])
footfall_df.info()
footfall_df.head()
```

```python
calendar_df['date'] = pd.to_datetime(calendar_df['date'])
calendar_df.describe()
calendar_df.info()
calendar_df.head()
calendar_df['event'] = calendar_df['event'].str.lower()
```

```python
calendar_df['is_exam'] = calendar_df['event'].str.contains('exam|quiz|midsem|endsem', regex=True).astype(int)
calendar_df['is_holiday'] = calendar_df['event'].str.contains('holiday|sunday|diwali|republic|independence|eid|labour', regex=True).astype(int)
calendar_df['is_weekend'] = calendar_df['date'].dt.weekday.isin([5, 6]).astype(int)  # 5=Sat, 6=Sun
calendar_df['is_fest'] = calendar_df['event'].str.contains('fest|carnival|inter iit|kshitij|techfest|sports meet|cultural|infi|umang', regex=True).astype(int)
exam_dates = calendar_df.loc[calendar_df['is_exam'] == 1, 'date'].sort_values().tolist()

#def days_to_next_exam(today):
#    future = [d for d in exam_dates if d >= today]
 #   return (future[0] - today).days if future else 30

#calendar_df['days_to_next_exam'] = calendar_df['date'].apply(days_to_next_exam)
```

```python
weather_df['timestamp'] = pd.to_datetime(weather_df['date'] + ' ' + weather_df['time'])
weather_df.drop([ 'date','time'], axis=1, inplace=True)
weather_df['timestamp']=pd.to_datetime(weather_df['timestamp'])
weather_df.describe()
weather_df.info()
weather_df.head()
```

```python
menu_df['date'] = pd.to_datetime(menu_df['date'])
menu_df.describe()
menu_df.info()
menu_df.head()
```

```python
footfall_df['date'] = footfall_df['timestamp'].dt.date
footfall_df['hour'] = footfall_df['timestamp'].dt.hour
evening = footfall_df[(footfall_df['hour'] >= 19) & (footfall_df['hour'] < 21)]
peak_queue = evening.groupby('date')['headcount'].max().reset_index()
peak_queue.columns = ['date', 'peak_queue']
peak_queue.head()
peak_queue.info()
```

```python
peak_queue = peak_queue.sort_values('date')
peak_queue['w_day_moving_avg'] = peak_queue['peak_queue'].rolling(window=5, min_periods=1).mean().shift(1)
```

```python
peak_queue['yesterday_peak_queue'] = peak_queue['peak_queue'].shift(1)
peak_queue['date'] = pd.to_datetime(peak_queue['date']).dt.date
peak_queue['date']=pd.to_datetime(peak_queue['date'])
peak_queue.head()
```

```python
menu_wide = menu_df.pivot(index='date', columns='meal', values='main_course').reset_index()
menu_wide.columns = ['date', 'breakfast_menu', 'dinner_menu', 'lunch_menu']
menu_wide.head()
```

```python
weather_df['hour'] = weather_df['timestamp'].dt.hour
weather_df['date'] = weather_df['timestamp'].dt.date

weather_till_4 = weather_df[weather_df['hour'] < 16]
weather_features = weather_till_4.groupby('date').agg({
    'temp_c': 'mean',
    'humidity_pct': 'max'
}).reset_index()
weather_features['date']=pd.to_datetime(weather_features['date'])
weather_features.head()
```

```python
df = pd.DataFrame()
menu_wide['date'] = pd.to_datetime(menu_wide['date'])
menu_wide= menu_wide.merge(calendar_df[['date', 'is_exam', 'is_holiday', 'is_weekend','is_fest']], on='date', how='left')
menu_wide = menu_wide.merge(weather_features, on='date', how='left')
menu_wide = menu_wide.merge(peak_queue, on='date', how='left')
```

```python
menu_wide.head()
```

```python
def contains(item, keyword):
    if isinstance(item, str):
        return int(keyword.lower() in item.lower())
    return 0

menu_wide['dinner_contains_paneer'] = menu_wide['dinner_menu'].apply(lambda x: contains(x, 'paneer'))
menu_wide['dinner_contains_chicken'] = menu_wide['dinner_menu'].apply(lambda x: contains(x, 'chicken'))
menu_wide['dinner_is_heavy'] = menu_wide['dinner_menu'].apply(lambda x: contains(x, 'biryani') or contains(x, 'chole') or contains(x, 'puri'))
```

```python
dinner_footfall = menu_wide.groupby('dinner_menu')['peak_queue'].mean()
popularity_map_dinner = dinner_footfall.to_dict()
dinner_menu = menu_wide[['date', 'dinner_menu']]
dinner_menu.columns = ['date', 'dinner_item']
dinner_menu['popularity_score_dinner'] = dinner_menu['dinner_item'].map(popularity_map_dinner)
menu_wide = menu_wide.merge(dinner_menu[['date', 'popularity_score_dinner']], on='date', how='left')
menu_wide.head()
```

```python
Q1 = menu_wide['peak_queue'].quantile(0.25)
Q3 = menu_wide['peak_queue'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
menu_wide_clean = menu_wide[(menu_wide['peak_queue'] >= lower_bound) & (menu_wide['peak_queue'] <= upper_bound)]
print(f"Removed {len(menu_wide) - len(menu_wide_clean)} outlier rows.")
menu_wide_clean.info()
```

```python
X = menu_wide_clean.drop(columns=['date', 'dinner_menu', 'breakfast_menu','lunch_menu','peak_queue','is_exam','humidity_pct','is_weekend','dinner_is_heavy','is_fest','is_holiday','dinner_contains_chicken','dinner_contains_paneer'])
Y = menu_wide_clean['peak_queue']
Y.info()
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

X_train, X_test, y_train, y_test = train_test_split(X, Y, shuffle=False, test_size=0.2)

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print("MAE:", mean_absolute_error(y_test, y_pred))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R²: {r2:.2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(y_test.values, label='Actual', marker='o')
plt.plot(y_pred, label='Predicted', marker='x')
plt.title("Actual vs Predicted Queue Peak")
plt.xlabel("Day")
plt.ylabel("Peak Queue (people/min)")
plt.legend()
plt.grid()
plt.show()
```

```python
errors = y_test - y_pred

plt.hist(errors, bins=20, edgecolor='black')
plt.title("Prediction Error Distribution")
plt.xlabel("Error (Actual - Predicted)")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

```python
results_df = X_test.copy()
results_df['Actual'] = y_test.values
results_df['Predicted'] = y_pred
results_df['Error'] = results_df['Actual'] - results_df['Predicted']
results_df['Absolute_Error'] = results_df['Error'].abs()

# Add back 'date' if it was preserved
if 'date' in df.columns:
    results_df['date'] = df.loc[X_test.index, 'date']

results_df.sort_values(by='Absolute_Error', ascending=False).head(10)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

# Define the param grid
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 15, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['sqrt', 'log2']
}

# Create the model
rf = RandomForestRegressor(random_state=42)

# Grid Search
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=3,  # 3-fold cross-validation
    scoring='neg_mean_absolute_error',
    verbose=1,
    n_jobs=-1
)

# Fit the grid search
grid_search.fit(X_train, y_train)

# Print best parameters
print("Best parameters:", grid_search.best_params_)

# Evaluate best model
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"Tuned MAE: {mae:.2f}")
print(f"Tuned RMSE: {rmse:.2f}")
print(f"Tuned R²: {r2:.2f}")
```
